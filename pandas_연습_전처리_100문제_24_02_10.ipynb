## Apply, Map function

```python
url ='https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv'

df = pd.read_csv(url,index_col=0)
df
```

```python
# income_Category의 카테고리를 map 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라
#  Unknown : N
#  Less than $40K : a
#  $40K - $60K : b
#  $60K - $80K : c
#  $80K - $120K : d
#  $120K +’ : e
```

```python
dic = {
    'Unknown' : 'N',
    'Less than $40K' : 'a' ,
    '$40K - $60K' : 'b',
    '$60K - $80K' : 'c',
    '$80K - $120K' : 'd',
    '$120K +': 'e'
}
df['new_income'] = df.Income_Category.map(lambda x: dic[x])
df['new_income']
```

```python
# Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라
# Unknown : N
# Less than $40K : a
# $40K - $60K : b
# $60K - $80K : c
# $80K - $120K : d
# $120K +’ : e

def changeCategory(x):
  if x =='Unknown':
    return 'N'
  elif x=='Less than $40K':
    return 'a'
  elif x=='$40K - $60K':
    return 'b'
  elif x=='$60K - $80K':
    return 'c'
  elif x=='$80K - $120K':
    return 'd'
  elif x=='120K +':
    return 'e'

df['newIncome'] = df.Income_Category.apply(changeCategory)
# df.drop('new_income',axis=1,inplace = True)
df
```

```python
# customer_Age 의 값을 이용하여 나이 구간을 AgeState 컬럼으로 정의하라
# [0~9 : 0 , 10~19: 10], 20~29:20 각 구간의 빈도수를 구하여라

df['AgeState'] = df.Customer_Age.map(lambda x: x//10 *10)
ans = df.AgeState.value_counts().sort_index()
ans
```

```python
# Education_level 의 값중 Graduate 단어가 포함되는 값은 1 그렇지 않은 경우에는 0으로 변경하여
# newEdulevel 컬럼을 정의하고 빈도수를 출력하라

df['newEdulevel'] = df['Education_Level'].map(lambda x: 1 if 'Graduate' in x else 0)
df.newEdulevel.value_counts()
```

```python
# Credit_Limit 컬럼값이 4500 이상인 경우 1 그외의 경우에는 모두 0으로 하는 newLimit 정의하라.
# newLimit 각 값들의 빈도수를 출력하라
df['newLimit'] = df['Credit_Limit'].map(lambda x:1 if x >=4500 else 0)
df['newLimit'].value_counts()
```

```python
# Marital_Status 컬럼값이 Married 이고 Card_Category 컬럼의 값이 Platinum인 경우 1
# newState의 각 값들의 빈도수를 출력하라 그외의 경우에는 모두 0으로 하는 newState컬럼을 정의하라.

def check(x):
  if (x.Marital_Status =='Married') and (x.Card_Category == 'Platinum'):
    return 1
  else:
    return 0
df['newState'] = df.apply(check,axis=1)
ans = df['newState'].value_counts()
ans
```

```python
#Gender 컬럼값 M인 경우 male
#F인 경우 female로 값을 변경하여 Gender 컬럼에 새롭게 정의하라. 각 value의 빈도를 출력하라


def Gender(x):
  if x.Gender == 'M':
    return 'male'
  else:
    return 'female'

df['Gender'] = df.apply(Gender,axis=1)
df['Gender'].value_counts()
```

```python
df =pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv')
df
```

```python
# 년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경하라.
# 서울시의 제공데이터의 경우 0시가 24시로 표현된다
# datetime 기본 제공은 년-월-일:시(00:00:00)이므로 이 형태로 만들어 준다.
import datetime
def change(x):
  hour = x.split(':')[1]
  date = x.split(':')[0]

  if hour=='24':
    hour='00:00:00'
    FinalDate =pd.to_datetime(date +''+hour) +datetime.timedelta(days=1)

  else:
    hour = hour +':00:00'
    FinalDate = pd.to_datetime(date+""+hour)
  return FinalDate

df['Date'] = df['Date'].apply(change)  # 년월시로 변경할 것 컬럼
ans = df
ans
```

```python
# 일자별 영어요일 이름을 dayName 컬럼에 저장하라
# df.rename(columns={'(년-월-일:시)':'Date'},inplace=True)  # 컬럼명이 괄호안에 있어서 타이핑오류가 심해서 변경
#df['Date'].dt.time.value_counts()  # 변경이 잘 된 것으로 확인

df['dayName'] = df['Date'].dt.day_name()
df['dayName']
```

```python
# 요일별 각 PM10등급의 빈도수를 파악하라
df.groupby(['dayName','PM10등급'],as_index=False).size().head()
```

```python
# 시간이 연속적으로 존재하며 결측치가 없는지 확인하라
# 시간을 차분했을 경우 첫 값은 Nan, 이후 모든 차분값이 동일하면 연속이라고 판단한다.

check = len(df['Date'].diff().unique())   # nan 과 특정 동일한 값 총 두개가 나온다! 첫번째 차분은 당연히 없기에 nan값에 할당된다.

if check ==2:
  ans = True
else:
  ans = False
ans
```

```python
# 오전 10시와 오후 10시(22시)의 PM10의 평균값을 각각 구하여라
# 시간대 별로 묶는다 .. 그럼 거기에서 iloc로 10시 , 22시를 행으로 가져오고 첫번째 컬럼만 선택한다
ans = df.groupby(df['Date'].dt.hour).mean().iloc[[10,22],[0]]
ans
```

```python
# 날짜 컬럼을 index로 만들어라
# ans = df.set_index('Date',inplace=True, drop=True)
df
```

```python
#데이터를 주단위로 뽑아서 최소,최대 평균, 표준표차를 구하여라

df.resample('W').agg(['min','max','mean','std'])
```

## merge , concat !!

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/mergeTEst.csv',index_col= 0)
df
```

```python
df1 = df.iloc[:4,:]
df2 = df.iloc[4:,:]

display(df1)
display(df2)
```

```python
# df1과 df2 데이터를 하나의 데이터 프레임으로 합쳐라
merged_df = pd.concat([df1,df2])
merged_df
```

```python
df3 = df.iloc[:2,:4]
df4 = df.iloc[5:,3:]

display(df3)
display(df4)
#df3과 df4 데이터를 하나의 데이터 프레임으로 합쳐라. 둘다 포함하고 있는 년도에 대해서만 고려한다

Ans = pd.concat([df3,df4],join='inner')
Ans
```

```python
# df3과 df4 데이터를 하나의 데이터 프레임으로 합쳐라. 모든 컬럼을 포함하고, 결측치는 0으로 대체한다
Ans = pd.concat([df3,df4],join='outer').fillna(0)
Ans
```

```python
# df5 = df.T.iloc[:7,:3]
# df6 = df.T.iloc[6:,2:5]

# display(df5)
display(df6)
#df5과 df6 데이터를 하나의 데이터 프레임으로 merge함수를 이용하여 합쳐라.
# Algeria컬럼을 key로 하고 두 데이터 모두 포함하는 데이터만 출력하라

Ans = pd.merge(df5,df6,on='Algeria',how='inner')
Ans
# inner로 했고 algeria를 기준으로 df5 와 df6 의 공통된것은 2016년도 뿐이다 그래서 2016년도에 value들이 호출 되었따.
```

```python
# df5과 df6 데이터를 하나의 데이터 프레임으로 merge함수를 이용하여 합쳐라. Algeria컬럼을 key로 하고 합집합으로 합쳐라
Ans =pd.merge(df5,df6,on='Algeria',how='outer')
Ans
```
